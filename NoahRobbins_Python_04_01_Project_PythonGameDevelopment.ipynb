```python

import random #imported random library to generate random colours 
from IPython.display import clear_output #imported IPython.display library to clear outputs later

#list of colours and dictionary of corresponding emojis
colours = ['red', 'white', 'purple', 'blue', 'orange', 'yellow']
dots = {'red': "\U0001F534", 'white': "\u26AA", 'purple': "\U0001F7E3", 'blue': "\U0001F535", 'orange': "\U0001F7E0", 'yellow': "\U0001F7E1"}

#Global variables to use as stats at the end
count = 0
incorrect_guess = []
avg_incorrect = 0

# Function to generate a random combination of colors
def generate_random():
    '''
    Generates random combination of colours using random library.

    Returns: list containing the four colours
    '''
    combination = []
    for i in range(4):
        random_colour = random.choice(colours)
        combination.append(random_colour)
    return combination

# Function to get user guess
def get_player_guess():
    '''
    Asks user to input their guess and validates it in lower case

    Returns: string of user's guess
    '''
    while True:
        guess = input("Your guess: ").lower()
        if guess in colours:
            return guess
        else:
            print("Incorrect input, please choose a colour from the options provided above.")

# Function to print all available colours
def print_all_colours():
    '''
    Prints all availale colours with corresponding emojis from dictionary.
    '''
    for key, value in dots.items():
        print(value, end=' ')

#prints introduction message and reminds users of the rules if they haven't read the instructions (README File)
print_all_colours(), print_all_colours(), print_all_colours(), print_all_colours()
print("\nWELCOME! Are you the Mastermind? Let's play to find out...")
print("Choose 4 colours. You will be shown symbols which hint at the answer after each round of guesses.")
print("A colour in the correct position:âœ…")
print("A correct colour in the wrong position:â˜‘")
print("No symbols indicate that none of your colours are correct")
print("Input ONE of the following colours: red, white, purple, blue, orange, yellow")

#Game loop begins - used a while loop so that the game will continue to run unless stopped at the end
while True:
    store = []  # Stores all guesses for the current game (within while loop)
    combination = generate_random()  # Calls the function to generate secret code 

    for i in range(8): #length of game is 8 guesses
        count += 1 #adds to global variable count
        if i >= 2:  # After 2 rounds of guessing the output is cleared so that player can focus on the current and previous output only
            clear_output(wait=True)  #calling the IPython.display library
            print("\nPrevious Guesses:") 
            print("Options: red, white, purple, blue, orange, yellow") #prints options again in case user forgets colours
            if len(store) > 0:  # if there are elements in the store then it will print the corresponding emoji for the guess
                for guess, hint in store:
                    print(' '.join(dots[color] for color in guess) + ' ' + ' '.join(hint))
        
        print("\nYou have {} guesses remaining".format(8 - i)) #prints number of guesses remaining

        all_guesses = []  #variable to contain the lists of guesses

        for x in range(4): #stores all the guesses in the list 
            guess = get_player_guess()
            all_guesses.append(guess)

        # counts all the colours in the correct positions (green ticks). Sum +1 for each time this is true (list comprehension)
        correct_positions = sum(1 for i in range(4) if all_guesses[i] == combination[i])
            

        # counts for correct colors in wrong positions (grey ticks)
        #only want to count each colour once so a min is used to prevent double counting
        correct_colors = sum(min(all_guesses.count(color), combination.count(color)) for color in colours) - correct_positions

        #green tick multiplied by number of correct answers and grey tick multiplies by correct colours in wrong place
        hint = ["âœ…"] * correct_positions + ["â˜‘"] * correct_colors

        store.append((all_guesses, hint))  # Add the current guess and its hint to the store for stacking later

        
        # Print the current guess and ticks
        print("Current Guess:", ' '.join(dots[color] for color in all_guesses))
        print("Hints:", ' '.join(hint))
        
        #Winning
        if correct_positions == 4:
            print("YOU WIN, YOU ARE THE MASTERMIND!!!")
            break
        else:
            for number, colour in enumerate(all_guesses): #counting the incorrect guesses by collecting all guesses and counting them individually (enumerate)
                if colour not in combination: #if statement to identify incorrect values
                    incorrect_guess.append((i + 1, colour)) #stored incorrect values
    #Losing
    else:
        print("You've run out of guesses! Game over. You are not the Mastermind.") 
    #Reveal the results to the loser
    print("Correct Sequence:", ' '.join(dots[color] for color in combination))
    
    #option to play again
    play_again = input("Do you want to play again? (yes/no): ").lower()
    if play_again == "no": #if no then the game loop is broken
        break
    if play_again == "yes": #if yes then the game will continue until they lose
        continue
    else:
        print("Please type yes or no") #ensure they only type yes or no

#calculates average incorrect guesses out of total guesses
avg_incorrect = len(incorrect_guess) / float(count)

#game stats
print("\nGame Stats")
print("Total Guesses:", count)
print("Incorrect Guesses:", len(incorrect_guess))
print("Average of incorrect guesses:", avg_incorrect)
```
